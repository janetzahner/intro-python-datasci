```python
#filtering
#filtering subsets cases using a conditional test
#this gets all manufacturers
planes_df['manufacturer']
```

```python
#this gets all manufacturers
#this gets just where manufacturer = EMBRAER
planes_df['manufacturer'] == 'EMBRAER'
```

```python
planes_df[planes_df['manufacturer'] == 'EMBRAER'].head()
```

```python
#build this so that the True values are the values we want
cond = planes_df['manufacturer'] == 'EMBRAER'
cond
```

```python
#only brings back the True values
planes_df[cond].head()
```

```python
planes_df.loc[planes_df['manufacturer']]
```

```python
planes_df.loc[planes_df['year'] > 2002].head()
```

```python
planes_df.loc[(planes_df['year'] > 2002) & (planes_df['year'] < 2004)].head()
```

```python
#combine the filter criteria in variables
cond_yr1 = planes_df['year'] > 2002
cond_yr2 = planes_df['year'] < 2004
cond_yr = cond_yr1 & cond_yr2
cond_yr #data
```

```python
sum(cond_yr) #total True values
```

```python
planes_df.loc[cond_yr1 & cond_yr2].head()
```

```python
#find planes that have more than 3 engines
cond_engines = planes_df['engines'] > 3
cond_engines
```

```python
planes_df.loc[cond_engines].head()
```

```python
sum(cond_engines)
```

```python
sum(planes_df['engines'] > 3)
```

```python
#filtering and slicing together
planes_df_filtered = planes_df.loc[planes_df['manufacturer'] == 'EMBRAER']
planes_df_filtered_and_selected = planes_df_filtered[['year', 'engines']]
planes_df_filtered_and_selected.head()
```

```python
#or you can do slicing and filtering in a single line
planes_df.loc[planes_df['manufacturer'] == 'EMBRAER', ['year', 'engines']].head()
```

```python
#this is a nice clean approach
rows = planes_df['manufacturer'] == 'EMBRAER'
cols = ['year', 'engines']
```

```python
planes_df.loc[rows, cols].head()
```

```python
#find entries from planes_df that have planes made by Boeing 
#and only seats and model variables
rows = planes_df['manufacturer'] == 'BOEING'
cols = ['seats', 'model']
```

```python
planes_df.loc[rows, cols].head()
```
