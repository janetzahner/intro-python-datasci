```python
import pandas as pd
planes = pd.read_csv('../data/planes.csv')
```

```python
planes.head()
```

```python
#total person capacity per plane, including the flight crew
seats = planes['seats']
seats.head()
```

```python
capacity = seats+5
capacity.head()
```

```python
#order doesn't matter
capacity_2 = 5 + seats
capacity_2.head()
```

```python
#add the capacity value into the data frame as a new column
#whatever is inside the brackets becomes the new column name
#must be explicitly named
planes['capacity'] = capacity
planes.head()
```

```python
# you can do this all in one step instead of using a variable
planes['capacity'] = planes['seats'] + 5
```

```python
planes.head()
```

```python
#add a new column called first_class_seats taht is 1/5 of total seats on the plane
planes['first_class_seats'] = planes['seats'] *.2
planes.head()
```

```python
#overwriting columns
#pretend we learned that engines is actually per wing... so plane has engines*2
#syntax is the same, just reference a column that already exists
planes['engines'] = planes['engines'] *2
planes.head()
```

```python
#new column based on multiple existing column values
seats_per_engine = planes['seats'] / planes['engines']
seats_per_engine.head()
```

```python
planes['seats_per_engine'] = seats_per_engine
planes.head()
```

```python
#nonsense column to practice
technology_index = ((planes['year']-1900) / 4) + (planes['engines'] * 2)
```

```python
planes['technology_index'] = technology_index
planes.head()
```
