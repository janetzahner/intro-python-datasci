```python
import pandas as pd
planes_df = pd.read_csv('../data/planes.csv')
```

```python
planes_df.head()
```

```python
#this doesn't actually return any data
planes_df.groupby('manufacturer')
```

```python
type(planes_df.groupby('manufacturer'))
```

```python
#define which summary stats you want
#the groupby field becomes the index
planes_df.groupby('manufacturer').agg({'seats': ['mean', 'max']}).head()
```

```python
df_groupby = planes_df.groupby('manufacturer').agg({'seats': ['mean', 'max']}).head()
df_groupby.loc['AIRBUS']
```

```python
#the groupby field becomes the index
df_groupby.index
```

```python
#if you don't want the groupby to become the index, add an extra clause
planes_df.groupby('manufacturer', as_index = False).agg({'seats': ['mean', 'max']}).head()
```

```python
#you can group by multiple columns
planes_df.groupby(['manufacturer', 'year'], as_index = False).agg({'seats': ['mean', 'max']}).head()
```

```python
#Fix the below code to find the minimum number of seats on a plane by year:
#added .head() to shorten result set
planes_df.groupby('year').agg({'seats': ['min']}).head()
```

```python
planes_df.groupby('year').agg({'seats': ['min']}).index
```

```python
#don't make year the index
planes_df.groupby('year', as_index = False).agg({'seats': ['min']}).head()
```

```python
#with sorting and not re-setting the index
planes_df.groupby('year', as_index = False).agg({'seats': ['min']}).sort_values(by = ('seats', 'min'), ascending=False)
```
