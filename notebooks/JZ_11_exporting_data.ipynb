```python
import pandas as pd
planes_df = pd.read_csv('../data/planes.csv')
planes_df.head()
```

```python
planes_df2 = planes_df[['year', 'manufacturer']]
planes_df2.head()
```

```python
df = planes_df2.groupby('manufacturer', as_index=False).agg({'year': 'mean'})
df.head()
```

```python
#write the data to a file
#you must specify a file extension
#it will overwrite an existing file if there's already one with this name
df.to_csv('../data/cleaned.csv')
```

```python
pd.read_csv('../data/cleaned.csv').head()
#"unnamed" is the index from the previous data frame
#don't need to save the default index since it will create a new one
```

```python
df.to_csv('../data/cleaned.csv', index=False)
pd.read_csv('../data/cleaned.csv').head()
#skips the index
#sep and header are good parameters to set too when exporting to csv
```
